```python
import os
import json
```

```python
def make_annotations_ids_unique(json_file_path):
    with open(json_file_path,'r') as f:
        coco_data = json.load(f)

    annotation_id = 1
    for annotation in coco_data['annotations']:
        annotation['id'] = annotation_id
        annotation_id += 1

    with open(json_file_path, 'w') as f:
        json.dump(coco_data,f,indent=4)
    print('UNIQUE ANNOTATION ID SAVE!')
```

```python
train_json_path = "/data/ephemeral/working_yolo/cocolabels/train/coco_annotations.json"
val_json_path = "/data/ephemeral/working_yolo/cocolabels/val/coco_annotations.json"
```

```python
make_annotations_ids_unique(train_json_path)
make_annotations_ids_unique(val_json_path)
```
